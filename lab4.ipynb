```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, classification_report
```

```python
df = pd.read_csv("UCI_Credit_Card.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Разделение на X и y
X = df.drop(["default.payment.next.month"], axis=1)
y = df["default.payment.next.month"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
# Инициализация Random Forest
rf_clf = RandomForestClassifier(
    n_estimators=200,
    criterion="gini",
    random_state=42,
    n_jobs=-1
)
```

```python
# Обучение модели
rf_clf.fit(X_train, y_train)
```

```python
# Предсказания
y_pred = rf_clf.predict(X_test)
```

```python
# Оценка
acc = accuracy_score(y_test, y_pred)
prec = precision_score(y_test, y_pred)
rec = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
roc = roc_auc_score(y_test, rf_clf.predict_proba(X_test)[:,1])
```

```python
print("Accuracy:", acc)
print("Precision:", prec)
print("Recall:", rec)
print("F1 Score:", f1)
print("ROC AUC:", roc)

print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score, roc_auc_score,
    mean_squared_error, mean_absolute_error, r2_score
)
```

```python
X_clf = df_credit.drop("default.payment.next.month", axis=1)
y_clf = df_credit["default.payment.next.month"]
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score, roc_auc_score,
    mean_absolute_error, mean_squared_error, r2_score
)
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from collections import Counter
```

```python
df_credit = pd.read_csv("UCI_Credit_Card.csv")
df_credit.head()
```

```python
X_clf = df_credit.drop("default.payment.next.month", axis=1)
y_clf = df_credit["default.payment.next.month"]
```

```python
X_clf_train, X_clf_test, y_clf_train, y_clf_test = train_test_split(
    X_clf, y_clf, test_size=0.25, random_state=42, stratify=y_clf
)
```

```python
df_air = pd.read_csv('AirQuality.csv', sep=';', decimal=',')

df_air = df_air.dropna(axis=1, how='all')

df_air = df_air.drop(columns=["Date", "Time"]).fillna(method="ffill")

df_air.replace(-200, np.nan, inplace=True)
df_air.dropna(inplace=True)

df_air.head()
```

```python
target = "CO(GT)"
X_reg = df_air.drop(columns=[target])
y_reg = df_air[target]
```

```python
X_reg_train, X_reg_test, y_reg_train, y_reg_test = train_test_split(
    X_reg, y_reg, test_size=0.25, random_state=42
)
```

```python
class RandomForestClassifierCustom:
    def __init__(self, n_estimators=100, max_depth=None, random_state=42):
        self.n_estimators = n_estimators
        self.max_depth = max_depth
        self.random_state = random_state
        self.trees = []
        self.rng = np.random.RandomState(random_state)

    def _bootstrap_sample(self, X, y):
        n_samples = X.shape[0]
        indices = self.rng.choice(n_samples, size=n_samples, replace=True)
        return X[indices], y[indices]

    def fit(self, X, y):
        self.trees = []
        for _ in range(self.n_estimators):
            X_sample, y_sample = self._bootstrap_sample(X, y)
            tree = DecisionTreeClassifier(
                max_depth=self.max_depth,
                random_state=self.rng.randint(0, 10000)
            )
            tree.fit(X_sample, y_sample)
            self.trees.append(tree)
        return self

    def predict(self, X):
        predictions = np.array([tree.predict(X) for tree in self.trees])
        final_preds = []
        for i in range(X.shape[0]):
            vote = Counter(predictions[:, i]).most_common(1)[0][0]
            final_preds.append(vote)
        return np.array(final_preds)
```

```python
class RandomForestRegressorCustom:
    def __init__(self, n_estimators=100, max_depth=None, random_state=42):
        self.n_estimators = n_estimators
        self.max_depth = max_depth
        self.random_state = random_state
        self.trees = []
        self.rng = np.random.RandomState(random_state)

    def _bootstrap_sample(self, X, y):
        n_samples = X.shape[0]
        indices = self.rng.choice(n_samples, size=n_samples, replace=True)
        return X[indices], y[indices]

    def fit(self, X, y):
        self.trees = []
        for _ in range(self.n_estimators):
            X_sample, y_sample = self._bootstrap_sample(X, y)
            tree = DecisionTreeRegressor(
                max_depth=self.max_depth,
                random_state=self.rng.randint(0, 10000)
            )
            tree.fit(X_sample, y_sample)
            self.trees.append(tree)
        return self

    def predict(self, X):
        predictions = np.array([tree.predict(X) for tree in self.trees])
        return predictions.mean(axis=0)
```

```python
rf_clf_custom = RandomForestClassifierCustom(n_estimators=100, max_depth=None, random_state=42)
rf_clf_custom.fit(X_clf_train.values, y_clf_train.values)
y_pred_clf_custom = rf_clf_custom.predict(X_clf_test.values)
```

```python
rf_reg_custom = RandomForestRegressorCustom(n_estimators=100, max_depth=None, random_state=42)
rf_reg_custom.fit(X_reg_train.values, y_reg_train.values)
y_pred_reg_custom = rf_reg_custom.predict(X_reg_test.values)
```

```python
print("Accuracy:", accuracy_score(y_clf_test, y_pred_clf_custom))
print("Precision:", precision_score(y_clf_test, y_pred_clf_custom))
print("Recall:", recall_score(y_clf_test, y_pred_clf_custom))
print("F1-score:", f1_score(y_clf_test, y_pred_clf_custom))
```

```python
print("MAE:", mean_absolute_error(y_reg_test, y_pred_reg_custom))
print("MSE:", mean_squared_error(y_reg_test, y_pred_reg_custom))
print("R2:", r2_score(y_reg_test, y_pred_reg_custom))
```

```python
rf_clf_custom_improved = RandomForestClassifierCustom(n_estimators=200, max_depth=10, random_state=42)
rf_clf_custom_improved.fit(X_clf_train.values, y_clf_train.values)
y_pred_clf_custom_improved = rf_clf_custom_improved.predict(X_clf_test.values)

rf_reg_custom_improved = RandomForestRegressorCustom(n_estimators=200, max_depth=10, random_state=42)
rf_reg_custom_improved.fit(X_reg_train.values, y_reg_train.values)
y_pred_reg_custom_improved = rf_reg_custom_improved.predict(X_reg_test.values)
```

```python
print("Accuracy:", accuracy_score(y_clf_test, y_pred_clf_custom_improved))
print("Precision:", precision_score(y_clf_test, y_pred_clf_custom_improved))
print("Recall:", recall_score(y_clf_test, y_pred_clf_custom_improved))
print("F1-score:", f1_score(y_clf_test, y_pred_clf_custom_improved))
```

```python
print("MAE:", mean_absolute_error(y_reg_test, y_pred_reg_custom_improved))
print("MSE:", mean_squared_error(y_reg_test, y_pred_reg_custom_improved))
print("R2:", r2_score(y_reg_test, y_pred_reg_custom_improved))
```
